```python
import pandas as pd
import numpy as np
import nltk
```

```python
train = pd.read_csv("data/train.csv")
test = pd.read_csv("data/test.csv")
```

```python
train.head()
```

```python
import string
from copy import deepcopy

punctuation = string.punctuation
stopwords = nltk.corpus.stopwords.words('english')

def process_sentence(essay):
    processed_essay = deepcopy(essay)
    punctation_signs = 0
    to_remove = []
    for char in essay:
        if char in punctuation:
            punctation_signs += 1
            to_remove.append(char)
    for char in to_remove:
        processed_essay = processed_essay.replace(char, " ")
    processed_essay = processed_essay.replace('\n', " ")
    processed_essay = processed_essay.lower()
    res = []
    for word in processed_essay.split(" "):
        if word not in stopwords and word!='':
            res.append(word)
    return res, punctation_signs
```

```python
processed_essay_col = []
punctuation_signs_col = []
for i in range(0, train.shape[0]):
    processed_essay, punctuation_signs = process_sentence(train.full_text[i])
    processed_essay_col.append(processed_essay)
    punctuation_signs_col.append(punctuation_signs)
train["processed_essay"] = processed_essay_col
train["punctuation_signs"] = punctuation_signs_col
```
